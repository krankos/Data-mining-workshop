```python
import os 
import pandas as pd
D = pd.read_excel('./data/Profils.xlsx', sheet_name=0, index_col=0)
D
```

```python
from sklearn import manifold

mds = manifold.MDS(n_components=2, dissimilarity="precomputed", random_state=1, normalized_stress=False)

mds.fit(D)
```

```python
points = mds.embedding_
points
```

```python
import matplotlib.pyplot as plt

ax = plt.axes([0., 0., 2., 2.])
ax.set_aspect('equal')

plt.scatter(points[:, 0], points[:, 1], color='turquoise', s=100, lw=0, label='MDS')

for label, x, y in zip(D.index, points[:, 0], points[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-20, 20),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

plt.show()
```

```python
mds.stress_
```

```python
# kruskal stress
import numpy as np
stress_kruskal = np.sqrt(mds.stress_ / (0.5 * (D**2).sum().sum()))
stress_kruskal
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=1)
kmeans.fit(points)

kmeans.cluster_centers_
```

```python
ax=plt.axes([0., 0., 1., 1.])
ax.set_aspect('equal')
plt.scatter(points[:, 0], points[:, 1], c=kmeans.labels_, cmap='rainbow', s=100, lw=0, label='MDS')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=100, lw=3, c='black', zorder=10)
```

```python
kmeans.labels_
```

```python
ax=plt.axes([0., 0., 1., 1.])
ax.set_aspect('equal')
plt.scatter(points[:, 0], points[:, 1], c=kmeans.labels_, cmap='rainbow', s=100, lw=0, label='MDS')
for i in range(len(points)):
    plt.annotate(D.index[i], (points[i, 0], points[i, 1]), color='black', weight='bold', fontsize=10, ha='center', va='center')
```

```python
from sklearn.metrics import silhouette_score

sil = np.zeros(10)

for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(points)
    sil[k] = silhouette_score(points, kmeans.labels_)

sil

```

```python
plt.plot(range(2, 10), sil[2:10], marker='o')
```

```python
segments = []

for i in range(D.shape[0]):
    for j in range(i+1, D.shape[0]):
        if (D.iloc[i,j]==0):
            segments.append([points[i,:], points[j,:]])

len(segments)
```

```python
from matplotlib.collections import LineCollection
lc = LineCollection(segments, linewidths=0.5, color='black', alpha=0.5)
ax = plt.axes([0., 0., 1., 1.])

ax.set_aspect('equal')

plt.scatter(points[:, 0], points[:, 1], c=kmeans.labels_, cmap='rainbow', s=100, lw=0, label='MDS')
for i in range(len(points)):
    plt.annotate(D.index[i], (points[i, 0], points[i, 1]), color='black', weight='bold', fontsize=10, ha='center', va='center')

ax.add_collection(lc)

```
